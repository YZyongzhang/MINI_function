```python
with open('blog.html' , 'r+' , encoding='utf-8') as file:
    page = file.read()
soup = BeautifulSoup(page , 'lxml')
# print(soup.prettify)
```

```python
div_contend = soup.find_all('div' , id='content_views')
print(div_contend)
```

```python
open("div.html" , "w+").write(str(div_contend))
```

```python
import urllib.request as url
from bs4 import BeautifulSoup
request = url.Request(url='https://blog.csdn.net/carefree2005/article/details/123006684')
html = url.urlopen(request).read().decode('utf-8')
# with open('blog.html' , 'w+' , encoding= 'utf-8') as f:
#     f.write(html)
html = BeautifulSoup(html , 'lxml')
```

```python
html
```

```python
div_content = html.find("div" , id = "content_views")
print(div_content)
```

```python

origin_html = open("myhtml.html" , 'r+')
origin_page = BeautifulSoup(origin_html , 'lxml')
target_div = origin_page('div' , class_ = "content")
target_div.append(div_content)
with open("finnal.html" , "w+") as file:
    file.write(origin_page.prettify())
```
